### 1) Realice y grafique una imagen con una cruz negra y fondo blanco.

```python
import numpy as np
import matplotlib.pyplot as plt

negro=np.zeros((80,80))
blanco=negro+1

fila=np.concatenate((blanco,negro,blanco),axis=1)

cruz=np.concatenate((fila, np.zeros((80,240)),fila),axis=0)

plt.imshow(cruz, cmap='gray')
plt.show()
print(cruz.shape)
```

### 9) Aplicar una máscara  de suavizado de 5x5 a la componente roja de una imagen a elección, un filtro de bordes a la componente verde componer una imagen a color donde se reemplace las componentes rojas y verdes modificadas.

```python
#Filtro gaussiano
import cv2
import numpy as np
import matplotlib.pyplot as plt
 
img = cv2.imread('paisajepuente.jpg')#carga la imagen como numpy de uint8
gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)#Convierte en escala de grises

#n=int(input('Ingrese tamaño matriz para filtro gaussiano (número impar y positivo) '))
n = 5

print(img.shape)
print(gris.shape)
print(img.max())
blur = cv2.GaussianBlur(gris,(n,n),0) #aca es donde se aplica el filtro Gaussiano

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

plt.subplot(121),plt.imshow(gris,cmap='gray',vmin=0,vmax=255),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur,cmap='gray',vmin=0,vmax=255),plt.title('Filtro Gaussiano')
plt.xticks([]), plt.yticks([])
#print(type(blur[0] [0]))#verifica que devuelve uint8
plt.show()

```

```python
# VER LO QUE ESTA EN NTB2
#Continúa el ejemplo anterior
#Mostrar las componentes verde y azul

import numpy as np
rojo = img[:,:,2]
verde = img[:,:,1]
azul = img[:,:,0]
    
    
monocroma=np.transpose(sum(np.transpose(img)/3)) #intensidad monocroma=(rojo+verde+azul)/3
print("dimensiones de imagen monocroma  (R+G+B)/3=" , monocroma.shape)

#para mostrar gráfico de paleta
l=np.linspace(0,ima_arr.max(),256)#vector de intensidades
l=l.astype('uint8')

tr=np.zeros((256,40),'uint8')
for i in range(40):
    tr[:,i]=l

paleta='inferno'

fig, axs = plt.subplots(nrows=2, ncols=3, sharex=True)
fig.set_figheight(10)
fig.set_figwidth(15)
axs[0,0].imshow(ima_arr)# ,cmap=paleta)
axs[0,0].set_title("imagen color RGB")
axs[0,1].imshow(monocroma ,cmap='gray')
axs[0,1].set_title("imagen monocroma traspuesta (R+G+B)/3")
axs[0,2].imshow(tr,cmap='gray')
axs[0,2].set_title(" referencia (paleta)")
axs[1,0].imshow(rojo ,cmap='Reds')
axs[1,0].set_title("componente roja:R=Reds")
axs[1,1].imshow(verde ,cmap='Greens')
axs[1,1].set_title("componente verde:G=Greens")
axs[1,2].imshow(azul ,cmap='Blues')
axs[1,2].set_title("componente azul:B=Blues")

plt.show()
```

```python
import cv2
satelital=cv2.imread('arboles.jpg')
satelital= cv2.resize(satelital,(618,426))
cv2.imshow('pulse una tecla aquí',satelital)
cv2.waitKey() #espera pulsado de tecla. 
cv2.destroyAllWindows()
```

```python
from matplotlib import pyplot as plt

plt.imshow(satelital[:,:,[2,1,0]])
print(satelital.shape)
plt.show()
```

```python
import numpy as np
dims=satelital.shape
cant_filas=dims[0]
cant_columnas=dims[1]
cant_pixeles=cant_filas*cant_columnas
data0=np.zeros([cant_pixeles,4],dtype='int')
for f in range(cant_filas):
    for c in range(cant_columnas):
        nro=f*cant_columnas+c
        data0[nro,0]=nro
        data0[nro,1:4]=satelital[f,c,:]

import pandas as pd
df = pd.DataFrame(data0, columns=["pixel_id","azul","verde","rojo"])

print(df.shape)
print('Base de pixeles:')
df.head(9)   
#print(satelital[0,0,1])
```

```python
showCrosshair = False
fromCenter = False


roi1 = cv2.selectROI("delimite la ROI+ent+tecla", satelital, fromCenter, showCrosshair)#ROI:region of interest

cv2.waitKey() #espera pulsado de tecla. 
print('ROI1:',roi1)
print('(columna, fila, anchocol, anchofila)')
cv2.destroyAllWindows()
# columna inicial,fila inicial,ancho columna,ancho fila del roi
```

```python
showCrosshair = False
fromCenter = False
roi2 = cv2.selectROI("delimite la ROI+ent+tecla", satelital, fromCenter, showCrosshair)
cv2.waitKey() #espera pulsado de tecla. 
print('ROI2:',roi2)
print('(columna, fila, anchocol, anchofila)')
cv2.destroyAllWindows()
```

```python
showCrosshair = False
fromCenter = False
roi3 = cv2.selectROI("delimite la ROI+ent+tecla", satelital, fromCenter, showCrosshair)
cv2.waitKey() #espera pulsado de tecla. 
print('ROI3:',roi3)
print('(columna, fila, anchocol, anchofila)')
cv2.destroyAllWindows()
```

```python
rois=[roi1,roi2,roi3]
```

```python
import numpy as np

bordes = cv2.Sobel(satelital,cv2.CV_64F,1,1,ksize=5)#capa de bordes ()

satelital2=np.zeros((bordes.shape[0],bordes.shape[1],6)) # defino una imagen con 6 canales

satelital2[:,:,0:3]=satelital #3 primeros canales imagen original
satelital2[:,:,3:6]=bordes # 3 últimos canales con bordes
print(satelital2.shape)
```

```python
import numpy as np
cantidad=0
for i in range(3):
    cantidad=cantidad+(rois[i][2]*rois[i][3])
print (cantidad)
data=np.zeros([cantidad,6])
target=np.zeros(cantidad)
cuenta=0
for i in range(3):
    roi=rois[i]
    
    for fil in range (roi[1],roi[1]+roi[3]):
        for col in range (roi[0],roi[0]+roi[2]):
            data[cuenta,:]=satelital2[fil,col,:]
            target[cuenta]=i
            cuenta=cuenta+1
    
print(cuenta)    
import pandas as pd
df = pd.DataFrame(data, columns=["azul","verde","rojo","bordes_azul","bordes_verde","bordes_rojo"])
print(df.shape) # la cantida de pixeles es la suma de los pixeles seleccionados cuando se recorta suelo,fondo,cultivo
print('target=',target, '; dimensión target=',target.shape)
df.head(10)
```

```python
from sklearn.neighbors import KNeighborsClassifier
n_neighbors = 5

knn = KNeighborsClassifier(n_neighbors) #se define el clasificador
knn.fit(data, target)#se lo entrena/ajusta con los datos de entrenamiento
print('Presición (Accuracy) del clasificador K-NN en el conjunto de entrenamiento: {:.2f}'
     .format(knn.score(data, target)))
print('Específicamente el {:.2f}'
     .format(100*knn.score(data, target)),'% de los datos de entrenamiento fueron clasificados correctamente')
```

```python
a=satelital.shape
print(a)
clasif=np.zeros(a[0:2])
print(knn.predict([[40, 80, 40, 0,0, 255]]))
for i in range(a[0]): #filas
    for j in range(a[1]): #columnas
        clasif[i,j]=knn.predict([satelital2[i,j,:]])
        
print('No se puede evaluar el porcentaje de acierto o Presición en el resto de la imagen')
print('Realizaremos una valoración visual')        
```
